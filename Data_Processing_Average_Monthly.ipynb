# Averages of DistilRoBERTa emotions by month

```python
# average —— distillroberta

import pandas as pd
df = pd.read_csv("DistilRoBERTa.csv")
df.columns = df.columns.str.strip()

df["publication_date"] = pd.to_datetime(df["publication_date"], errors="coerce")
df = df.dropna(subset=["publication_date"]).copy()

emotion_cols = ["fear", "anger", "disgust", "joy", "sadness", "surprise"]

for col in emotion_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df["month"] = df["publication_date"].dt.strftime("%Y-%m")

monthly_emo = (
    df.groupby("month")[emotion_cols]
      .mean()
      .reset_index()
)

article_count = (
    df.groupby("month")
      .size()
      .reset_index(name="article_count")
)

monthly_emo = monthly_emo.merge(article_count, on="month", how="left")

monthly_emo.to_csv(
    "monthly_average_distilroberta.csv",
    index=False
)

print(monthly_emo.head())
print(monthly_emo["month"].head())
```

# Averages of LLM emotions by month

```python
#average-LLM

import pandas as pd
import numpy as np


input_file = "LLM_merged_all_articles.csv"
output_file = "LLM_monthly_emotion_rounded.csv"

df = pd.read_csv(input_file)

emotion_cols = ["Fear", "Anger", "Disgust", "Joy", "Sadness", "Surprise"]


df["publication_date"] = pd.to_datetime(df["publication_date"], errors="coerce")

df["month"] = df["publication_date"].dt.to_period("M").astype(str)

monthly_df = df.groupby("month", as_index=False)[emotion_cols].mean()

for col in emotion_cols:
    monthly_df[col] = monthly_df[col].apply(
        lambda x: pd.NA if pd.isna(x) else int(np.floor(x + 0.5))
    ).astype("Int64")

monthly_df["month_sort"] = pd.to_datetime(monthly_df["month"], format="%Y-%m", errors="coerce")
monthly_df = monthly_df.sort_values("month_sort").drop(columns="month_sort")

monthly_df.to_csv(output_file, index=False, encoding="utf-8-sig")

print("Saved to:", output_file)
print(monthly_df.head())
```
