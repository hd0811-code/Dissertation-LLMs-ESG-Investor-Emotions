# DistilRoBERTa

```python
from transformers import pipeline
import pandas as pd
import torch
classifier = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    top_k=None,
    truncation=True,
    max_length=512,
    device=0 if torch.cuda.is_available() else -1
)
```

```python
df = pd.read_csv("all_articles_title_dedup.csv")
```

```python
df['publication_date'] = pd.to_datetime(df['publication_date'])
```

```python
def get_emotion_scores(text):
    if pd.isna(text):
        return None
    
    result = classifier(str(text))
    return {item['label']: item['score'] for item in result[0]}
```

```python
emotion_scores = df['text'].apply(get_emotion_scores)
emotion_df = pd.json_normalize(emotion_scores)

df = pd.concat([df, emotion_df], axis=1)
```
